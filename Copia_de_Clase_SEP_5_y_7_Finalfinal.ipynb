```python
# cuantos jugadores les falta informacion de birth state? cuantos la tienen?

#len(nba.query("birth_state.notnull()"))  #Forma como la hizo el Profe de mostrar Todos los que tienen Info

tienen=nba['birth_state'].str.contains('').value_counts()

Falta= len(nba)-nba['birth_state'].str.contains('').value_counts()

print('Jugadores que tienen Data : ',tienen)
print('Jugadores que no tienen data en el campo : ',Falta)
```

```python
# cual es la altura media de jugadores de University of Kentucky?

nba.query("collage == 'University of Kentucky'")['height'].mean()
```

```python
#Para crear una columna nueva en el data set.

nba['Columna nueva Edad']=2023 - nba['born'].to_frame()
nba.head(2)
```

```python
# calcular IMC de cada jugador (peso / altura ** 2, altura en metros, peso en kg y la Edad)

nba['Edad']=2023-nba['born']
nba['IMC']=(nba['weight']/nba['height'])**2
nba['Altura en Metros']=nba['height']/100



nba.head(2)
```

```python
# quien es el jugador con IMC mas alto y bajo de la historia de la NBA

Alto=nba['IMC'].sort_values().max()
Bajo=nba['IMC'].sort_values().min()

print('Imc Max = ',Alto,'  / Imc Mínimo = ',Bajo)
```

```python
#CUALES JUGADORES  MIDE MAS DE 2 METOS
nba.query("height >200")
```

```python
#quien es el jugador con IMC mas alto
nba.sort_values(by='IMC').query("Player.notnull()").tail(1)#['Player'].values[0]
```

```python
#Cuál es la Altura Medía de los jugadores por Universidad

nba.groupby("collage")['height'].mean()
```

```python
#Mostrar lista de jugadores cuyo peso sea SUperior a 100

nba.query("weight > 100")
```

```python
# encontrar el collage con average weight mas alto

nba.groupby(["collage"], as_index=False)['weight'].mean().sort_values(by='weight').tail(1)
```

```python
# encontrar el collage con median weight mas alto

nba.groupby(['collage'],as_index=False)['weight'].median().sort_values(by='weight').tail(1)
```

```python
# encontrar todos los collages con altura media mayor que 210

nba.groupby(["collage"],as_index=False)['height'].mean().query("height > 210")
```

```python
# dentro de los college con altura media mayor que 190, cual es el que tiene peso medio mas bajo?

nba.groupby(['collage'],as_index=False)['height','weight'].mean().query("height > 190").sort_values(by='weight').head(1)
```

```python
# FORMA 2 DEL EJERSICIO ANTERIOR COMO LO HIZO EL PROFESOR CON MÉTODO AGG
# dentro de los college con altura media mayor que 190, cual es el que tiene peso medio mas bajo?

nba.groupby('collage',as_index=False).agg({'height':'mean','weight':'mean'}).query("height > 190").sort_values(by='weight').head(1)#['collage'].values[0]
```

```python
# encuentra todos los college con altura media mayor que 190 y peso medio mayor que 100

nba.groupby('collage',as_index=False).agg({'height':'mean','weight':'mean'}).query("height > 190").query('weight > 100')
```

```python
# dentro de los jugadores de New York, cual es el college con altura media mas alta?

nba.query("birth_state == 'New York'").groupby('collage',as_index=False)['height'].mean().sort_values(by='height').tail(1)#.values[0]
```

```python
# dentro de los jugadores de New York, cual es el college con mas jugadores de la nba?

nba.query("birth_state == 'New York'")['collage'].value_counts()#.head(1)
```

```python
# cual es la pareja de college / birth state con mas jugadores?

nba.groupby(['collage', 'birth_state'], as_index=False).size().sort_values("size")

#.size(): Después de agrupar los datos, se aplica el método size() para contar el número de filas en cada grupo.
#Esto generará un nuevo DataFrame con las columnas 'collage', 'birth_state' y 'size', donde 'size' representa la cantidad de
# jugadores en cada grupo (es decir, cuántos jugadores asistieron a la misma universidad y nacieron en el mismo estado)
```

```python
# Mostrar la media de las alturas,la mediana de los pesos y el maximo de nacimiento por Colleage
 # Para renombrar Columnas Calculadas se Utilisa .rename

nba.groupby(['collage'],as_index=False).agg({'height':'mean','weight':'median','born':'max'}).rename(columns={'height': 'height_mean(PROMEDIO)','weight':'Weight_Median(MEDIANA)','born':'born_max(MÁXIMO)'})

```
