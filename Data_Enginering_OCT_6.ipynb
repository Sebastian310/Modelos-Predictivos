```python
# 7_ Qué proporción de jugadores de la NBA estudiarion en la universidad de Kentucky?

import pandas as pd
import numpy as np

nba=pd.read_csv("nba.csv")

Kentucky= len(nba.query("collage == 'University of Kentucky'"))
rest= len(nba["Player"])

result=Kentucky/rest
print(" El : ",round(result * 100,2),"% Del total de los jugadores de la NBA estudiaron en la Universidad de kentuchy")
```

```python
#11_ Dibuja un histograma de la altura de los jugadores de la NBA.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

nba = pd.read_csv("nba.csv")

sns.histplot(data=nba, x='height', bins=20, kde=True, color='green')
plt.title('Histograma Suavizado de la Altura de Jugadores de la NBA')
plt.xlabel('Altura (pulgadas)')
plt.ylabel('Frecuencia')
plt.grid(False)
```

```python
#11_ Dibuja un histograma de la altura de los jugadores de la NBA. (SIn Suavizar)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

nba = pd.read_csv("nba.csv")

plt.hist(nba['height'], bins=20, edgecolor='k', alpha=0.7)
plt.title('Histograma de la Altura de Jugadores de la NBA')
plt.xlabel('Altura (pulgadas)')
plt.ylabel('Frecuencia')
plt.grid(False)
plt.show()
```

```python
#12_ Dibuja un diagrama de linea o de puntos con la altura media por año de nacimiento.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

nba = pd.read_csv("nba.csv")

altura_media_por_año = nba.groupby('born')['height'].mean().reset_index() #Ahondar en Reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=altura_media_por_año, x='born', y='height', marker='*')
plt.title('Altura Media por Año de Nacimiento en la NBA')
plt.xlabel('Año de Nacimiento')
plt.ylabel('Altura Media')
plt.grid(True)
plt.show()
```

```python
#13_ Dibuja un diagrama de barras con la altura media por estado de nacimiento.

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

nba = pd.read_csv("nba.csv")

altura_media_x_estado = nba.groupby('birth_state')['height'].mean().reset_index() #Reset Index otra vcez

plt.figure(figsize=(15, 6))
sns.barplot(data=altura_media_x_estado, x='birth_state', y='height', palette='viridis') # que es palette
plt.title('Altura Media por Estado de Nacimiento en la NBA')
plt.xlabel('Estado de Nacimiento')
plt.ylabel('Altura Media (pulgadas)')
plt.xticks(rotation=90)
plt.show()
```

```python
#14_ Dibuja un diagrama de barras con la altura media por estado de nacimiento, solo para
#    los estados con mas de 100 jugadores

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

nba = pd.read_csv("nba.csv")

jugadores_por_estado = nba['birth_state'].value_counts()

estados_con_mas_de_100_jugadores = jugadores_por_estado[jugadores_por_estado > 100].index

nba_filtrado = nba[nba['birth_state'].isin(estados_con_mas_de_100_jugadores)]

altura_media_por_estado = nba_filtrado.groupby('birth_state')['height'].mean().reset_index()


plt.figure(figsize=(12, 6))
sns.barplot(data=altura_media_por_estado, x='birth_state', y='height', palette='viridis')
plt.title('Altura Media por Estado de Nacimiento en la NBA (Estados con más de 100 jugadores)')
plt.xlabel('Estado de Nacimiento')
plt.ylabel('Altura Media (pulgadas)')
plt.xticks(rotation=90)
plt.show()
```
